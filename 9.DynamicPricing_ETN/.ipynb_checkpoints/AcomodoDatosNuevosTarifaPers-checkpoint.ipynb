```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import os
import json

import Tools4Cluster as TC
from src.dynamic_pricing_data_loader import cargar_y_preparar_datos
```

```python
def GetDB():
    DB = pd.read_excel("ClusteringClientes.xlsx")
    return DB
```

```python
def GetTodayData():
    Df = TC.GetData()
    Df['FECHA_OPERACION'] = pd.to_datetime(Df['FECHA_OPERACION'])
    fecha_maxima = Df['FECHA_OPERACION'].max()
    Df = Df[Df['FECHA_OPERACION'] == fecha_maxima].copy()
    Df.loc[(Df['PORCENT_PROMO'] > 0) & (Df['DESC_DESCUENTO'] == 'ADULTO'), 'DESC_DESCUENTO'] = 'PROMOCION ESPECIAL'
    # Si tiene 0% de promoción pero estaba marcado como "PROMOCION ESPECIAL", devolverlo a "ADULTO"
    Df.loc[(Df['PORCENT_PROMO'] == 0) & (Df['DESC_DESCUENTO'] == 'PROMOCION ESPECIAL'), 'DESC_DESCUENTO'] = 'ADULTO'
    Df['VENTA_ANTICIPADA']= np.where(Df['VENTA_ANTICIPADA']=='SI',1,0)
    return Df
```

```python
def Preparacion():
    Df= GetTodayData()
    DB= GetDB()

```

```python
df_ = pd.DataFrame()
```

```python
# 1. Definimos las columnas de ORIGEN tal como las quieres, incluyendo 'VENTA' dos veces.
# Pandas creará un DataFrame con dos columnas idénticas llamadas 'VENTA' (o renombrará una internamente).
columnas_origen = [
    'BOLETOS_VEND',
    'VENTA', # Primera columna VENTA
    'VENTA', # Segunda columna VENTA
    'PORCENT_PROMO',
    'HORAS_ANTICIPACION',
    'VENTA_ANTICIPADA'
]

# 2. Definimos las columnas de DESTINO. Asegúrate de que este slice (DB.columns[1:7])
# contenga 6 nombres de columna que se alineen con el orden de las columnas de origen.
columnas_destino = DB.columns[1:7]

# 3. Asignación robusta usando .loc
# Se asigna el DataFrame de 6 columnas (RHS) a las 6 columnas de destino (LHS).
df_.loc[:, columnas_destino] = Df_Not[columnas_origen].values
df_['Recencia']=0
```

```python
df_
```

```python
df_dummies_desc = pd.get_dummies(Df_Not['DESC_DESCUENTO'], prefix='DESC').astype(int)
df_dummies_pago = pd.get_dummies(Df_Not['PAGO_METODO'], prefix='PAGO').astype(int)
#Df['EMAIL'],
componentes_alineados = [
    Df_Not['EMAIL'].reset_index(drop=True),
    df_.reset_index(drop=True),
    df_dummies_pago.reset_index(drop=True),
    df_dummies_desc.reset_index(drop=True)
]

# 2. Concatenamos sin usar ignore_index=True, lo que PRESERVA los nombres de columna.
# Como todos los componentes tienen ahora el mismo índice (0, 1, 2, ...), no habrá NaNs.
df_dummies_final = pd.concat(
    componentes_alineados,
    axis=1  # Concatenar como columnas
)
```
